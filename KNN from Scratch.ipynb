# KNN REGRESSION FROM SCRATCH

```python
# Function for knn-regression
def KnnReg(X_train, X_test, y_train, y_test,K=3):

    train=pd.concat([X_train,y_train],axis=1)
    test=pd.concat([X_test,y_test],axis=1)
    train.shape,test.shape
    test=test.set_index(keys=np.arange(0,len(test)))
    train=train.set_index(keys=np.arange(0,len(train)))
    dist=[]
    t=0
    k=K

    # finding the distance of test record from train records
    for i in range(test.shape[0]):
        distances_for_i = []
        for j in range(train.shape[0]):
            total=0
            for c in range(train.shape[1]-1):
                total+=(test.iloc[i,c]-train.iloc[j,c])**2
            d=np.sqrt(total)
            distances_for_i.append(d)
        dist.append(distances_for_i)

    # selecting y having  k min distances

    predictions = []
    for i in range(len(dist)):
        distances_for_i = dist[i]
        indices = np.argsort(distances_for_i)[:k]
        near_y_labels = [train.iloc[j, -1] for j in indices]
        y_hat = np.mean(near_y_labels)
        predictions.append(y_hat)

    # concating predicted value into the test data
    X_test['y_pred']=predictions
    return X_test
```

```python
KnnReg(X_train, X_test, y_train, y_test,3)
```
